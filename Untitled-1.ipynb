```python
operacao = str(input("Qual o tipo de operação utilizada? "))
num1 = float(input("Digite um número: "))
num2 = float(input("Digite outro número: "))
valor_adicao = num1 + num2
valor_subtracao = num1 - num2
valor_multiplicacao = num1 * num2

if operacao == 'adição':
    print(valor_adicao)

elif operacao == 'subtração':
    print(valor_subtracao)
elif operacao == 'multiplicação':
    print(valor_multiplicacao)
else:
    if operacao == 'divisão':
        if num2 == 0:
            print("Não é possível dividir por 0")
        else:
            print(num1 / num2)
    else:
        print("Operação inválida")
```
